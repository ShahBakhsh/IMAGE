```python
import yfinance
stocks = [ 'BTC-USD', 'ETH-USD', 'DOGE-USD' , 'MATIC-USD']


data = yfinance.download(stocks,
                         '2020-01-01',
                       '2022-01-01') ['Close']
```

```python
data.head()
```

```python
retruns = data.pct_change()
retruns.head()
```

```python
retruns = retruns. dropna()
data.head()
```

```python
average_daily_retruns = retruns.mean()
print (average_daily_retruns)
```

```python
standard_deviation_daily_retruns = retruns.std()
print (standard_deviation_daily_retruns)
```

```python
import numpy
weights = numpy.array ([0.4,0.2,0.3,0.1])
```

```python
covariance_matrix = (retruns.cov())*250
print (covariance_matrix )
```

```python
expected_portfolio_performance = numpy.sum (average_daily_retruns * weights)
print (expected_portfolio_performance)
```

```python
retruns ['Portfolio Retruns'] = retruns.dot(weights) # In Python NumPy dot() function is used to return the dot product of given arrays.
```

```python
retruns.head()
```

```python
daily_cumunlative_retruns = (1+retruns).cumprod()    #this represent cummulative product on dataset
print (daily_cumunlative_retruns)
```
